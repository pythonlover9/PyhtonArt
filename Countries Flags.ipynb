```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Indonesia
a = np.full((2,3,3),255)
a[0,:]=[255,0,0]
plt.imshow(a)
```

```python
#Ireland
a = np.full((3,6,3),255)
a[:,0:2]=[22,155,98]
a[:,4:6]=[255,136,62]
plt.imshow(a)
```

```python
a = np.array([[[22,155,98], [255,255,255], [255,136,62]]])
a = np.repeat(a,2,axis=1)
a = np.repeat(a,3,axis=0)

plt.imshow(a)
```

```python
#UAE
a = np.full((6,12,3),0)
a[:2]=[0,115,47]
a[2:4]=[255,255,255]
a[:,0:3]=[255,0,0]
plt.imshow(a)
```

```python
a = np.array([[[0,115,47]], [[255,255,255]], [[0,0,0]]])
a = np.repeat(a,2,axis=0)
a = np.repeat(a,9,axis=1)

b = [[[255,0,0]]]
b = np.repeat(b,6,axis=0)
b = np.repeat(b,3,axis=1)

c = np.hstack((b,a))
plt.imshow(c)
```

```python
H, W = 200,300
a    = np.full((H,W,3),(255,255,255))
Y, X = np.indices((H,W))
mask = Y >= H/2
a[mask] = (255,0,0)
mask = abs( Y - H/2 ) <= -2*X/3 + H/2
a[mask] = (0,0,255)
plt.imshow(a, interpolation='bilinear')
```

```python
H,  W  = 200,900
R, R1, R2 = 84, 10, 88

s = np.full((H,W,3),(255,153,31))                               #Saffron

w = np.full((H,W,3),(255,255,255))                              #White

I, J = np.indices((H,W))                                        #Ashoka Chakra
Y, X = -(I-H/2), J-W/2
r, phi = np.sqrt(X**2+Y**2), np.arctan2(Y, X)
mask = (r<R1) | (r<=R) & (np.cos(phi*24)>0) | (r>R) & (r<R2)
w[mask] = (0,0,128)

g = np.full((H,W,3),(19,136,8))                                 #Green

f = np.vstack((s,w,g))                                          #Vertical Stacking all of them

plt.imshow(f, interpolation='bilinear')
```
